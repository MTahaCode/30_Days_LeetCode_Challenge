```python
from flask import Flask, jsonify
from flask_cors import CORS
from pyngrok import ngrok
import os
import base64

import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Set up Flask app
app = Flask(__name__)
CORS(app)

@app.route('/')
def home():
    return jsonify(message="Hello from Flask!")

def encode_image(image_path):
    with open(image_path, "rb") as img_file:
        return base64.b64encode(img_file.read()).decode('utf-8')

@app.route("/get-image/<prompt>", methods=["GET"])
def get_image(prompt):
    # Replace with your actual image file path

    os.environ['HF_HUB_DISABLE_SYMLINKS_WARNING'] = '1'

    pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16)
    pipe = pipe.to("cuda")

    image = pipe(
        prompt= prompt,
        negative_prompt="",
        num_inference_steps=28,
        guidance_scale=7.0,
    ).images[0]

    # Save the image
    image.save("output_image.png")

    image_path = '/content/output_image.png'

    # Encode image to base64
    base64_image = encode_image(image_path)

    # Construct JSON response
    data = {
        'imageBase64': base64_image,
        'description': 'Optional image description'
    }

    return jsonify(data)

# Set up Ngrok with auth token and create the tunnel
ngrok.set_auth_token("2idYgrIM636GxyHUNo12pcMLoVL_2XsAa2RHcJ2oY498BWjiv")
public_url = ngrok.connect(5000)
print(f" * Ngrok tunnel \"{public_url}\" -> \"http://127.0.0.1:5000\"")

# Update the app's base URL (optional)
app.config["BASE_URL"] = public_url

# Run Flask app
app.run(host='0.0.0.0', port=5000)
```
